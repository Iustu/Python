```python
import yfinance
import matplotlib.pyplot as plt
import pyautogui
import time
import pyperclip
```

##  Busca

```python
codigo = input("Digite ação:")
dados = yfinance.Ticker(codigo).history("6mo")

fechamento = dados.Close
fechamento.plot()
```

## Análises

```python
maximo = fechamento.max()
minimo = fechamento.min()
atual = fechamento.iloc[-1]
```

```python
print(maximo)
print(minimo)
print(atual)
```

## Envio de email

```python
pyautogui.PAUSE = 0.3
pyautogui.press("win")
pyautogui.write("brave")
pyautogui.press("enter")


pyautogui.write("gmail.com")
pyautogui.press("ENTER")
time.sleep(3)

pyautogui.click(x=94, y=218)

pyautogui.PAUSE = 1

pyperclip.copy("Email.@gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.press("ENTER")
pyautogui.press("TAB")

pyperclip.copy('Análise diária.')
pyautogui.hotkey("ctrl","v")
pyautogui.press("TAB")

pyperclip.copy(f""" 
    Segue as análises pedidas da ação {codigo}.
    
    Cotação Máxima: R$ {round(maximo,2)}
    Cotação Mínima: R$ {round(minimo,2)}
    Cotação atual:  R$ {round(atual,2)}

    """)

pyautogui.hotkey("ctrl","v")

pyautogui.click(x=500, y=1010)
```
